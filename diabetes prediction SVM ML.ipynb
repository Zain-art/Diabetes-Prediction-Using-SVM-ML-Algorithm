# MAke the Prediction on new values comming 

```python
input_data=(1,70,30,90,70,11.8,0.191,3)
# make input_data as an array convert in numpy using asarray method
input_data_as_array=np.asarray(input_data)
# maker reshape of the input_data
input_data_reshape=input_data_as_array.reshape(1,-1)
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
# standardize the input value new 
scaler.fit(input_data_reshape)
std_data=scaler.transform(input_data_reshape)
print('standard data :',std_data)
prediction=classifier.predict(std_data)
print ('prediction new values: ',prediction)

if (prediction[0]==0):
    print('the person is not diabetes')
else:
    print('the person is diabetes')
```
